```python
import requests
from bs4 import BeautifulSoup
import json

# URL de la page à scraper
url = "https://www.trustpilot.com/review/www.teslamotors.com"

# Effectuer une requête HTTP GET
response = requests.get(url)

# Parser le contenu HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Trouver le script contenant les données JSON
script_tag = soup.find('script', {'id': '__NEXT_DATA__'})

if script_tag:
    # Charger le contenu JSON
    json_data = json.loads(script_tag.string)
    
    # Accéder à `totalPages`
    total_pages = json_data['props']['pageProps']['filters']['pagination']['totalPages']
    print(f"Le nombre total de pages est : \n{json.dumps(total_pages, indent=4)}")
else:
    print("Impossible de trouver les données nécessaires.")
```
